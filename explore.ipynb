```python
import pandas as pd
import os
```

```python
data_folder = "data/"
files = [f for f in os.listdir(data_folder) if f.endswith(".csv") ]
```

```python
for filename in sorted(files):
    df = pd.read_csv(os.path.join(data_folder, filename))
    columns = df.columns

    print(f"Filename: {filename}")

    print("\n=== SHAPE ===")
    print(df.shape)

    print("\n=== DTYPES ===")
    print(df.dtypes)
    
    print("\n=== COLUMNS ===")
    print(columns)
    
    print("\nColumn and Its Type")
    
    cols_and_types = ""
    
    for col in columns:
        col_and_type = f"{col} ({df[col].dtype})"
        if df.columns.get_loc(col) != df.shape[1] - 1:
            col_and_type += ", "
        else: 
            col_and_type += "."
            
        cols_and_types += col_and_type
    
    print(cols_and_types)
    
    print("\n=== 10 First Data ===")
    print(df.head(10))

    print("\n=== NULL COUNT ===")
    print(df.isna().sum())

    print("\n=== DUPLICATE COUNT ===")
    print(df.duplicated().sum())
    
    print("\n==========================\n")
```
