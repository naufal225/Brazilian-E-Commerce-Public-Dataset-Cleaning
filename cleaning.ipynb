```python
def clean_unicode(text):
    if not isinstance(text, str):
        return text
    
    text = text.lower().strip()
    
    cleaned_text = ''.join(c for c in unicodedata.normalize('NFD', text) if unicodedata.category(c) != 'Mn')
    
    return cleaned_text
```

```python
outlier_config = {
    "Order Items": {
        "dataframe": df_order_items,
        "columns": {
            "price": "is_price_outlier",
            "freight_value": "is_freight_value_outlier"
        }
    },
    "Order Payments": {
        "dataframe": df_order_payments,
        "columns": {
            "payment_value": "is_payment_value_outlier"
        }
    }
}

for name, config in outlier_config.items():
    
    df = config['dataframe']
    
    for col, flag in config['columns'].items():
        
        q1 = df[col].quantile(0.25)
        q3 = df[col].quantile(0.75)
        
        iqr = q3 - q1
        
        bottom_limit = q1 - 1.5 * iqr
        top_limit = q3 + 1.5 * iqr
        
        df[flag] = (df[col] < bottom_limit) | (df[col] > top_limit)
        
        total_outlier = df[flag].sum()
        presentage = (total_outlier / len(df) * 100).round(2)
        
        print(f"\n==={name}===\n")

        print(f"Total outlier: {total_outlier} ({presentage}%)")
        
        print(f"Q1: {q1}")
        print(f"Q3: {q3}")
        print(f"IQR: {iqr}")
        print(f"Bottom limit: {bottom_limit}")
        print(f"Top limit: {top_limit}\n")
        
        print(f"Jumlah yang 0 atau kurang dari 0 : {df[df[col] <= 0].shape[0]}")
        print(f"Jumlah lebih dari outlier        : {df[df[col] > top_limit].shape[0]}")
        print(f"Jumlah kurang dari outlier       : {df[df[col] < bottom_limit].shape[0]}\n")
        
        print(df.head(20))    
```
