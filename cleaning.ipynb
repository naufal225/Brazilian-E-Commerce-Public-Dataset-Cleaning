```python
def clean_unicode(text):
    if not isinstance(text, str):
        return text
    
    text = text.lower().strip()
    
    cleaned_text = ''.join(c for c in unicodedata.normalize('NFD', text) if unicodedata.category(c) != 'Mn')
    
    return cleaned_text
```

```python
df_geo['geolocation_city'] = df_geo['geolocation_city'].str.lower()

df_geo['geolocation_city'] = df_geo['geolocation_city'].str.strip()
```

```python
length_mean = df_products.groupby('product_category_name')['product_length_cm'].transform('mean')
height_mean = df_products.groupby('product_category_name')['product_heigth_cm'].transform('mean')
width_mean = df_products.groupby('product_category_name')['product_width_cm'].transform('mean')

df_products['product_length_cm'] = df_products['product_length_cm'].fillna(length_mean)
df_products['product_heigth_cm'] = df_products['product_heigth_cm'].fillna(height_mean)
df_products['product_width_cm'] = df_products['product_width_cm'].fillna(width_mean)
```
